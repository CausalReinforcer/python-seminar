fmin: Uses a Nelder-Mead simplex algorithm to find the minimum of function of
one or more variables

```python
from scipy.optimize import fmin

def model(par,x,y):
    ## some average offset plus a sin
    v = par[0] + par[1]*np.sin(2*np.pi*x/par[2] + par[3])
    return ((v - y)**2).sum()

res = fmin(model,np.array([yy.mean(),10,.1,1.0]),args=(tt,yy))
```

```python
mdl = res[0] + res[1]*np.sin(2*np.pi*tt/res[2] + res[3])
plot(tt,mdl)
plot(tt,yy)
```

```python
p=res[2]
t = tt % p
s = t.argsort()
scatter(t[s],mdl[s])
scatter(t[s],yy[s])
```
