$\textbf{Classifying NIST Handwritten Digits}$  
======================================================

We will try to classify handwritten digits (0-9) from their raw pixelated images.

Each image is 8x8 pixels.  We will not do any feature extraction and instead classify based on the intensity values for each pixel.

```python
%pylab inline
from sklearn import datasets, metrics
# import NIST digits data set (1797 8x8 images)
digits = datasets.load_digits()

plt.figure(figsize=(16, 6))
for i in range(10):
    plt.subplot(1,10,i)
    plt.imshow(digits['images'][i],cmap='gray',axes=None,interpolation='nearest')
    
```

$\textbf{Now we split the data into a training and testing set.}$

$\textbf{We will only fit the classifier on the training set and use the testing set to evaluate performance.}$

```python
# take the first 500 as training
train = 500
Xtr = digits['data'][:train]
Ytr = digits['target'][:train]
print("training size: " + str(len(Ytr)))

# testing set
Xte = digits['data'][train:]
Yte = digits['target'][train:]
print("testing size: " + str(len(Yte)))
```

Fit a Support Vector Machine Classifier
====

```python
# Create a classifier: a support vector machine classifier
from sklearn import svm
# instantiate classifier object
classifier = svm.SVC(gamma=0.005) # default: one-versus-one

# fit the classification model on training set
classifier.fit(Xtr, Ytr)

# make predictions for testing set
pred_svm = classifier.predict(Xte) 

print("True Class / Predicted class")
print(vstack((Yte[0:10],pred_svm[0:10])))
```

```python
# compute zero-one loss / score & confusion matrix
from sklearn import metrics

svm_01 = metrics.zero_one(Yte, pred_svm) # zero-one loss
svm_01_score = metrics.zero_one_score(Yte, pred_svm) # zero-one score
svm_confmat = metrics.confusion_matrix(Yte, pred_svm) # conf mat

print("Zero-One Loss: " + str(svm_01))
print("Zero-One Score: " + str(svm_01_score))
print("Confusion Matrix:")
print("[i, j] is the # of objects truly in group i but predicted to be in group j")
print(svm_confmat)
```

```python
# plot some digits that we get wrong
wrong = np.where(pred_svm != Yte)[0]
plt.figure(figsize=(15, 4))
for i in range(6):
    plt.subplot(1,6,i)
    plt.imshow(reshape(Xte[wrong[i]],(8,8)),cmap='gray',axes=None,interpolation='nearest')
    title("True = " + str(int(Yte[wrong[i]])) +". Pred = " + str(int(pred_svm[wrong[i]])))
```

```python
# compute precision and recall
# Note: precision & recall are for 2-class; multi-class returns weighted avg. prec/recall

svm_precision = metrics.precision_score(Yte, pred_svm) # TP / (TP + FP)
svm_recall = metrics.recall_score(Yte, pred_svm) # TP / (TP + FN)

print("Avg. Precision: ",svm_precision)
print("Avg. Recall: ", svm_recall)
```

```python
# compare to one-versus-all SVM
from sklearn import multiclass

# we can make predicitons in 1 line of code!
pred_svm_ovr = multiclass.OneVsRestClassifier(svm.SVC(gamma=0.005)).fit(Xtr, Ytr).predict(Xte)

print("0-1 Classification Scores:")
print("One-versus-One SVM: " + str(round(svm_01_score,3)))
print("One-versus-Rest SVM: " + str(round(metrics.zero_one_score(Yte,pred_svm_ovr),3)))
```

```python
# fit probabilistic SVM (via Platt Scaling)
svm_prob = svm.SVC(gamma=0.005,probability=True)
svm_prob.fit(Xtr, Ytr)
pred_svm_prob = svm_prob.predict_proba(Xte)

print("P(digit = 2) for first 6 testing set objects:")
print(Yte[0:6])
print(pred_svm_prob[0:6,2])
```

```python
#  ROC curve, AUC for SVM classifier using digit = 1
Yte_1 = map(lambda x: x == 1. and 1. or 0.,Yte) # does Y = 1
pred_svm_prob_1 = pred_svm_prob[:,1]
fpr, tpr, thresholds = metrics.roc_curve(Yte_1, pred_svm_prob_1)

plt.figure(figsize=(10, 8))
plt.plot(fpr,tpr,'b-',linewidth=3)
plt.xlim([0.,0.2])
plt.ylim([0.8,1.])
xlabel("False Positive Rate",size=15)
ylabel("True Positive Rate",size=15)
title("ROC Curve for NIST digit=1 SVM Classifier",size=22)
print("AUC for digit=1: " + str(metrics.auc(fpr,tpr)))
```

$\textbf{Tuning the Classifier}$  
======================================================

For now we have used a SVM classifier with the arbitrary choice of $\gamma = 0.005$ with the radial basis function kernel. 

The parameter $\gamma$  is the kernel bandwidth.

$\textbf{Let's see how we can rigorously find the optimal model using cross-validation and grid search.}$

```python
from sklearn import grid_search

# explore 2 different kernels and 6 choices of gamma
parameters = {'kernel':('linear', 'rbf'), 'gamma':[0.1, 0.01, 0.005, 0.001, 0.0005, 0.0001]}
svm_tune = grid_search.GridSearchCV(svm.SVC(), parameters, score_func=metrics.zero_one_score,\
                                    n_jobs = 2, cv = 5)
svm_opt = svm_tune.fit(Xtr, Ytr)
print("Best zero-one score: " + str(svm_opt.best_score_) + "\n")
print("Optimal Model:\n" + str(svm_opt.best_estimator_))
```

```python
# Find the best Random Forest classifier
from sklearn.ensemble import RandomForestClassifier

# explore 3 different forest sizes and 3 choices of mtry
parameters = {'n_estimators':[20,50,100],  'max_features':[8,10]}
rf_tune = grid_search.GridSearchCV(RandomForestClassifier(), parameters,\
                                   score_func=metrics.zero_one_score, n_jobs = 2, cv = 5)
rf_opt = rf_tune.fit(Xtr, Ytr)

print("Best zero-one score: " + str(rf_opt.best_score) + "\n")
print("Optimal Model:\n" + str(rf_opt.best_estimator_))
```

## Comparison of Random Forests (sklearn vs. R)

```python
# using sklearn RF
from sklearn.ensemble import RandomForestClassifier
def fitrf(Xtr,Ytr,ntree,mtry,Xte,Yte):
    rf1 = RandomForestClassifier(n_estimators=ntree,max_features=mtry)
    rf1.fit(Xtr, Ytr)
    preds = rf1.predict(Xte)
    print float(sum(preds!=Yte))/len(Yte)
    return preds

%timeit fitrf(Xtr,Ytr,1000,8,Xte,Yte)
```

$\textbf{Manifold Learning Example}$  
======================================================

The Swiss Roll data set

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
import pylab as pl

n_points = 3000
X, t = datasets.samples_generator.make_swiss_roll(n_samples=n_points)
fig = plt.figure(figsize=(12, 16))
ax = fig.add_subplot(211, projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2], c=t, cmap=pl.cm.Spectral)

ax1 = fig.add_subplot(212)
ax1.scatter(X[:,0], X[:,2], c=t, cmap=pl.cm.Spectral)
```

```python
from sklearn import manifold
out_dim = 2
n_neighbors = 10

Y = manifold.LocallyLinearEmbedding(n_neighbors, out_dim, eigen_solver='auto',\
                                    method='standard').fit_transform(X)
plt.figure(figsize=(10, 8))
plt.scatter(Y[:, 0], Y[:, 1], c=t, cmap=pl.cm.Spectral)
plt.title("LLE",size=25)
```

```python
Y = manifold.LocallyLinearEmbedding(n_neighbors, out_dim, eigen_solver='auto',\
                                    method='ltsa').fit_transform(X)
plt.figure(figsize=(10, 8))
plt.scatter(Y[:, 0], Y[:, 1], c=t, cmap=pl.cm.Spectral)
plt.title("LTSA",size=25)
```

```python
Y = manifold.Isomap(n_neighbors, out_dim).fit_transform(X)

plt.figure(figsize=(10, 8))
plt.scatter(Y[:, 0], Y[:, 1], c=t, cmap=pl.cm.Spectral)
plt.title("IsoMap",size=25)
```
