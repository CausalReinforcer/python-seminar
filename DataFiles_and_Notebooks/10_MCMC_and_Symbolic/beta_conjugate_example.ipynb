## Obama Approval Example - beta conjugacy

```python
%pylab inline
import numpy as np

n = 100
X = 55

alpha = 10
beta = 10
```

```python
from scipy import stats
x_theta = np.linspace(0, 1, 101)
```

```python
p_theta = stats.beta(alpha, beta).pdf(x_theta)

plot(x_theta, p_theta, linewidth=3.)
ylim([0,max(p_theta)])
xlabel("Theta")
ylabel("p(Theta)")
title("Prior Probability for Obama Approval Rating")
```

```python
alpha_post = alpha + X
beta_post = n - X + beta
post_theta = stats.beta(alpha_post, beta_post).pdf(x_theta)

plot(x_theta, post_theta, linewidth=3.,label='posterior')
plot(x_theta, p_theta, linewidth=1., label='prior')
ylim([0,max(post_theta)])
xlabel("Theta")
ylabel("p(Theta | X)")
title("Posterior Probability for Obama Approval Rating")
axvline(x=float(X)/n, linestyle='--', color='red', label="MLE")
legend()
```
