$\textbf{Bayesian Inference for the Mining Disaster Data Set}$  
======================================================

```python
%pylab inline
import DisasterModel
from pymc import MCMC

# instantiate MCMC object to handle our model
M = MCMC(DisasterModel)
```

We can visualize the model as a graphical model
======

this allows us to check that the model is written as desired (must have pydot and GraphViz installed)

```python
import pydot
import pymc.graph
pymc.graph.graph(M,format='png',path='',name='graph_mine',prog='dot')

import scipy.misc
figure(figsize=(8, 8))
imshow(scipy.misc.imread('graph_mine.png'))
```

```python
# starting value of parameters
print 'current value of s: ' + str(DisasterModel.s.value)
print 'current value of e: ' + str(DisasterModel.e.value)
print 'current value of l: ' + str(DisasterModel.l.value)
#DisasterModel.r.value
```

We can draw samples from the posterior using sample, 
======================================================
but we first need to 'burn in' the MCMC until it converges to a stationary distribution
and thin the chain to overcome auto-correlations in the samples

```python
# Run an MCMC chain of length 10,000 with 1000 burn-in and thinning by 10
# using Metropolis-Hastings algorithm
M.sample(iter=10000, burn=1000, thin=2)

# isample does interactive MCMC fitting, sample does non-interactive fitting
```

```python
# we can obtain summary statistics from the MCMC easily
M.stats() #summary statistics
```

```python
# plot trace of the sampler
plt.figure(figsize=(15, 8))
plt.plot(M.trace('s')[:])
```

```python
# plot posterior density for l
plt.figure(figsize=(12, 10))
plt.hist(M.trace('l')[:],normed=True,bins=20)
plt.xlabel('l (disasters per year)',fontsize=20)
plt.ylabel('Posterior Density',fontsize=20)

# plot posterior mean
lpostmean = np.mean(M.trace('l')[:])
plt.axvline(x=lpostmean, color='red',lw=5)

lpost2_5 = M.stats()['l']['quantiles'][2.5]
lpost97_5 = M.stats()['l']['quantiles'][97.5]
plt.axvline(x=lpost2_5, color='red',lw=3,linestyle='--')
plt.axvline(x=lpost97_5, color='red',lw=3, linestyle='--')
```

```python
# PyMC has its own plotting functionality, via the Matplot module
from pymc import Matplot

# plot function takes the model (or a single parameter) as an argument:
Matplot.plot(M)
```

```python
plt.figure(figsize=(12, 10))
plt.plot(M.trace('e')[:],M.trace('s')[:],'or',alpha=.3)
plt.xlabel('e (disasters per year)',fontsize=20)
plt.ylabel('s (changepoint + 1850)',fontsize=20)
```

Of course, what we really want is to make inference about the parameters of interest!
======

```python
# posterior estimation and inference for the year of changepoint

# posterior mean of year changepoint
print 'Posterior mean year for changepoint:  ' + str(round(1850 + M.stats()['s']['mean']))
# posterior 95% credible interval for s
print '95% credible interval for s:  (' + str(round(1850 + M.stats()['s']['quantiles'][2.5])) + \
',' + str(round(1850 + M.stats()['s']['quantiles'][97.5])) + ')'
```

```python
# posterior mean of magnitude of change (e - l)
eminusl = M.trace('e')[:] - M.trace('l')[:]
print 'Posterior mean magnitude of change in rate:  ' + \
str(round(np.mean(eminusl),2)) + ' disasters per year'
# posterior 95% credible interval for s
print '95% credible interval for change in rate:  (' + \
str(round(sort(eminusl)[np.floor(0.025*len(eminusl))],2)) + ',' + \
str(round(sort(eminusl)[np.floor(0.975*len(eminusl))],2)) + ')'
```

```python
# and we can also plot realizations of the model from the posterior!
dates = np.arange(1850,1961)
plt.figure(figsize=(12, 10))
plt.plot(dates,DisasterModel.disasters_array,'o-')
plt.xlim([1850,1960])
plt.xlabel('Year',fontsize=20)
plt.ylabel('# of Disasters',fontsize=20)
```

```python
# now we can plot realizations of the model from the posterior!
dates = np.arange(1850,1961)
plt.figure(figsize=(12, 10))
plt.plot(dates,DisasterModel.disasters_array,'o-')
plt.xlim([1850,1960])
plt.xlabel('Year',fontsize=20)
plt.ylabel('# of Disasters',fontsize=20)

# plot first 250 MCMC samples
for i in np.arange(250):
    ri = M.trace('r')[i]
    plt.plot(dates,ri,'-r',alpha=0.3)
```

Convergence Diagnostics
===

Geweke score = $\frac{\bar{\theta}_e - \bar{\theta}_b}{\sqrt{Var(\theta_e) + Var(\theta_b)}}$

```python
from pymc import geweke
from pymc.Matplot import geweke_plot
scores = geweke(M)
geweke_plot(scores)
```
