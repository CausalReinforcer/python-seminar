```python
%pylab inline
per_conv = lambda s: float(s.split("%")[0])/100.0
#print per_conv("5%")
data = loadtxt(open("astro.csv","r"),skiprows=1,delimiter=",",
        converters = {0: datestr2num, 2: per_conv, 4: per_conv})
```

```python
from scipy.signal import medfilt
t = data[:,0] - min(data[:,0])
errorbar(t,data[:,1],data[:,1]*data[:,2])
ys = medfilt(data[:,1],3)
plot(t,ys)
```

```python
from scipy.signal.spectral import lombscargle
max_days = 450.
min_days = 250.
nout  = 10000
freqs = np.linspace(1./max_days, 1/min_days, nout)
```

```python
t = data[:,0] - min(data[:,0])
normval = t.shape[0]
t.shape
pgram = lombscargle(t,ys[:],freqs)
pgram.shape
plot(freqs, np.sqrt(4*(pgram/normval)))
```
